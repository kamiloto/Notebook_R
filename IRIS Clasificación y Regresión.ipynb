## 3. Clasificación en diferentes especies usando el método del "K-Nearest Neighbor"

En este caso queremos ver si podemos predecir correctamente la clase de especie a la que pertenece una flor en funciónde las claracterísticas que tenemos dispinibles en la data: ancho del pétalo, la longitud del pétalo, la ancho del sépalo y la longitud del sépalo.

### 3.1. Partición de la data

Primero, dividimos el conjunto de datos Iris en entrenamiento y en el conjunto de datos de prueba para aplicar la clasificación KNN. El 70% de los datos se utiliza para la capacitación, mientras que la clasificación KNN se prueba en el 70% restante de los datos.

```r
set.seed(1234)
setosa = rbind(iris[iris$Species=="setosa",])
versicolor = rbind(iris[iris$Species=="versicolor",])
virginica = rbind(iris[iris$Species=="virginica",])


ind <- sample(1:nrow(setosa), nrow(setosa)*0.7)

iris.train<- rbind(setosa[ind,], versicolor[ind,], virginica[ind,])
iris.test<- rbind(setosa[-ind,], versicolor[-ind,], virginica[-ind,])
```

### 3.2. Elegir el valor óptimo del K

Ejecutamos el algoritmo de clasificación de KNN para diferentes valores de K y vemos el valor de K de K = 1 a K = 15, que da el error más bajo.

```r
error <- c()
for (i in 1:10)
{
  knn.fit <- knn(train = iris.train[,1:4], test = iris.test[,1:4], cl = iris.train$Species, k = i)
  error[i] = 1- mean(knn.fit == iris.test$Species)
}

ggplot(data = data.frame(error), aes(x = 1:10, y = error)) +
  geom_line(color = "Blue")
```

### 3.3. Precisión de predicción utilizando el valor K óptimo

Corremos el modelo de clasificación **KNN** en el conjunto de datos utilizando el valor de K como **K = 3**. Luego se verifica la tasa de clasificación errónea en las predicciones realizadas.

```r
set.seed(123)
iris_pred <- knn(train = iris.train[,1:4], test = iris.test[,1:4], cl = iris.train$Species, k=3)
table(iris.test$Species,iris_pred)
```

**Nota:** La tasa de clasificación errónea es **4.44%** (2 de 45 observaciones, estas 45 observaciones corresponden al 30% de los datos que se dejaron para la validación). Por lo tanto, se puede afirmar queel rendimiento del modelo de  clasificación **KNN** es bueno en el conjunto de datos de prueba.

### 4. Regresión

### Gráfico de dispersión

```r
plot(iris$Sepal.Length ~ iris$Sepal.Length, data = iris, xlab = "Longitud del sépalo", ylab = "Longitud del pétalo", 
     main ="Gráfico de dispersión")
```

### Estimación de la línea de regresión

Ahora dibujamos una línea de regresión calculada a partir de la suma de cuadrados mínimos usando la función **abline** aplicada a un objetivo de clase **lm (lineal model)**. Este objeto de clase lm consiste en una lista de 12 elementos que describen el modelo lineal.

```r
# Calcular los parámetros del modelo lineal
mod_ml <- lm(Petal.Length ~ Sepal.Length, data = iris)
mod_ml
```

```r
# Grafica de la línea de regresión estimada sobre el gráfico de dispersión
plot(Petal.Length ~ Sepal.Length, data = iris, xlab = "Longitud del sépalo", ylab = "Longitud del pétalo",  
     main = "Grafica de la línea de regresión estimada ")
     abline(mod_ml, col = "red", lwd = 2)
```

```r
# Vemos los resultados estadísticos del modelo de regresión lineal
summary(mod_ml)
```

Con base en los resultados del modelo **Mod_ml** se puede construir la ecuación linea de la longitud del pétalo usando los estimados de los coeficientes β del intercepto y la longitud del śepalo:


$$Petal.Length = −7.10144+1.85843 (Sepal.Length)$$

El valor de **R2=0.76** indica que el 76% de la variabilidad en la longitud del pétalo puede ser **explicada** por la variabilidad en la longitud del sépalo. Vemos que la asociación entre la longitud del pétalo y la longitud del sépalo es estadísticamente significantiva, con valor P<2e−16.

### 4. Verificación de supuestos

Los modelos de regresión lineal deben cumplir con 3 condiciones sin las cuales los estadísticos estimados y la ecuación lineal no son confiables.

##### I) Asociación lineal entre las variables

Se verifica visualmente que una línea recta representa la mejor interpretación de la relación entre las variables usando el gráfico de dispersión anteriormente. Asimismo, un gráfico de dispersión de los residuales sobre la variable predictora nos ayudaría a determinar si la relación es lineal.

```r
plot(mod_ml$residuals ~ iris$Petal.Length, xlab = "Longitud del pétalo", ylab = "Residuales")
abline(h = 0, col = "red", lwd = 3, lty = 3)
```

Una relación lineal será evidencia por una distribución simétrica de los residuales alrededor de la línea roja.

#### II) Normalidad de los residuales

Para verificar si los residuales siguen una distribución normal se puede utilizar un histograma.

```r
hist(mod_ml$residuals, xlab = "Residuales", ylab = "Frecuencia", main = "Normalidad de los residuales")
```

```r
# Un gráfico Q-Q también sería muy útil para determinar si los residuales siguen una distribución normal:

qqnorm(mod_ml$residuals, xlab = "Cuantiles teóricos", ylab = "Cuantiles muestrales")
qqline(mod_ml$residuals, lwd = 2, lty = 2, col = "red")
```
